Она должна включать в себя минимально следующие отношения:

    - студенты (минимальные атрибуты: ФИО, номер зачётки, дата рождения, группа, год поступления)
    - посещаемые курсы (минимальные атрибуты: название курса, тип курса, дата введения в программу)
    - преподаватели (минимальные атрибуты: ФИО, дата рождения, дата начала работы в нашем университете, дата начала трудовой деятельности)
Атрибуты отношений должны быть осмысленными и соответствовать тематике.

```python
conn = sqlite3.connect('university_database.db') 
c = conn.cursor()
```

```python
c.execute('''
          CREATE TABLE IF NOT EXISTS Students(
              [student_id] INTEGER PRIMARY KEY,
              [surname] TEXT,
              [name] TEXT,
              [lastname] TEXT,
              [credit_no] INTEGER,
              [birthday] DATE,
              [study_group] TEXT,
              [admissionYear] INTEGER) ''')

c.execute('''          
          CREATE TABLE IF NOT EXISTS Courses(
              [course_id] INTEGER PRIMARY KEY,
              [title] TEXT,
              [type] TEXT,
              [introductionDate] DATE) ''')

c.execute('''          
          CREATE TABLE IF NOT EXISTS Professors(
              [professor_id] INTEGER PRIMARY KEY,
              [surname] TEXT,
              [name] TEXT,
              [lastName] TEXT,
              [birthdate] DATE,
              [startDateUniversity] DATE,
              [startDateWork] DATE)''')

c.execute('''          
          CREATE TABLE IF NOT EXISTS Teaching(
              [professor_id] INTEGER,
              [course_id] INTEGER,
              FOREIGN KEY(professor_id) REFERENCES Professors(professor_id),
              FOREIGN KEY(course_id) REFERENCES Courses(course_id)
          )
          ''')

c.execute('''          
          CREATE TABLE IF NOT EXISTS Attending(
              [student_id] INTEGER,
              [course_id] INTEGER,
              FOREIGN KEY (student_id) REFERENCES Students(student_id),
              FOREIGN KEY (course_id) REFERENCES Courses(course_id))''')
```

Между сущностями есть отношения:
    1) профессор читает лекции (мноие ко многим)
    2) студент посещает курс (многие ко многим)
Поэтому пользуемся ассоциотивным отношением, остальное тривиально

```python
c.execute('''
        select title, cnt from
        (select course_id, count(*) as cnt
        from Teaching t
        group by course_id
        order by cnt limit 10) as tmp
        inner join Courses c on tmp.course_id=c.course_id
''')
pd.DataFrame(c.fetchall(), columns=['название курса', 'число преподов'])
```

Задание 4 [2 балла]. Напишите запрос, который выведет топ-10 преподавателей, которые преподают у наибольшего числа студентов. Расположите их в порядка убывания числа студентов.

```python
c.execute('''
        select surname, name, lastname, cnt from
        (select t.professor_id, count(*) as cnt
        from Teaching t
        join Attending a on t.course_id=a.course_id
        group by t.professor_id
        order by cnt desc limit 10) as tmp
        inner join Professors p on p.professor_id=tmp.professor_id
        
''')
pd.DataFrame(c.fetchall(), columns=['Фамилия', 'Имя', 'Отчество', 'Количество студентов'])
```

```python
c.execute('''
        select study_group, JULIANDAY(max(birthday)) - JULIANDAY(min(birthday))
        from Students
        where admissionYear >= 2019
        group by study_group
''')
pd.DataFrame(c.fetchall(), columns=['Группа', 'Разница между макс и мин датами рождения'])
```

Задание 6 [1.5 балла]. Среди преподавателей, родившихся в период между самой ранней датой начала работы в нашем университете и самой ранней датой начала трудовой деятельности по всем преподавателям, определите самое популярное имя.

```python
c.execute('''
        select name, count(*) as cnt
        from
        (select surname, name, lastname,
        (min(startDateUniversity) over() - strftime('%Y', birthdate)) as diff_with_uni,
        (strftime('%Y', birthdate) - min(startDateWork) over()) as diff_with_work
        from Professors) as tmp
        where diff_with_uni>=0 and diff_with_work>=0
        group by name
        order by cnt limit 1
        
''')
pd.DataFrame(c.fetchall(), columns=['Самое популярное имя среди искомых', 'Количество'])
```
